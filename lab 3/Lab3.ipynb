```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress

file_path = 'Lab3.2.csv' 
data = pd.read_csv(file_path)

# Data preparation
x1 = 1 / data['Sample']  
y1 = 1 / data['S1']  
yerr1 = (1/(data['S1']**2))*data['S1U']

x2 = 1 / data['Sample']  
y2 = 1 / data['S2']
yerr2 = (1/(data['S2']**2))*data['S2U']  

x3 = 1 / data['Sample']  
y3 = 1 / data['S3']
yerr3 = (1/(data['S3']**2))*data['S3U']

x4 = 1 / data['Sample']  
y4 = 1 / data['S4']
yerr4 = (1/(data['S4']**2))*data['S4U'] 


slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x1, y1)
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(x2, y2)
slope3, intercept3, r_value3, p_value3, std_err3 = linregress(x3, y3)
slope4, intercept4, r_value4, p_value4, std_err4 = linregress(x4, y4)


plt.figure(figsize=(10, 8))


plt.errorbar(x1, y1, yerr=yerr1, fmt='.', color='blue', capsize=5, label='Series 1')
plt.errorbar(x2, y2, yerr=yerr2, fmt='.', color='red', capsize=5, label='Series 2')
plt.errorbar(x3, y3, yerr=yerr3, fmt='.', color='pink', capsize=5, label='Series 3')
plt.errorbar(x4, y4, yerr=yerr4, fmt='.', color='green', capsize=5, label='Series 4')


plt.plot(x1, intercept1 + slope1*x1, color='blue', linestyle='--')
plt.plot(x2, intercept2 + slope2*x2, color='red', linestyle='--')
plt.plot(x3, intercept3 + slope3*x3, color='pink', linestyle='--')
plt.plot(x4, intercept4 + slope4*x4, color='green', linestyle='--')


plt.ylabel('1/Count Rate')
plt.xlabel('1/x')
plt.title('1/Count Rate vs. 1/x with Linear Regression')
plt.legend()
plt.grid(True)


plt.show()


print("Shelf 1: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope1, intercept1, r_value1**2))
print("Shelf 2: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope2, intercept2, r_value2**2))
print("Shelf 3: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope3, intercept3, r_value3**2))
print("Shelf 4: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope4, intercept4, r_value4**2))
print("U1=", std_err1)
print("U2=", std_err2)
print("U3=", std_err3)
print("U4=", std_err4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress

file_path = 'Lab3.2.csv' 
data = pd.read_csv(file_path)


x1 = data['Sample']  
y1 = np.log(data['S1']/data['Sample'])  #S1 is the data recorded on shelf 1 for each sample, sample is the sample # (1 through 5)
yerr1 = (1/(data['S1']))*data['S1U'] 

x2 = data['Sample']  
y2 = np.log(data['S2']/data['Sample'])  
yerr2 = (1/(data['S2']))*data['S2U']  

x3 = data['Sample']  
y3 = np.log(data['S3']/data['Sample'])  
yerr3 = (1/(data['S3']))*data['S3U']  

x4 = data['Sample']  
y4 = np.log(data['S4']/data['Sample'])  
yerr4 = (1/(data['S4']))*data['S4U']  


slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x1, y1)
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(x2, y2)
slope3, intercept3, r_value3, p_value3, std_err3 = linregress(x3, y3)
slope4, intercept4, r_value4, p_value4, std_err4 = linregress(x4, y4)


plt.figure(figsize=(10, 8))


plt.errorbar(x1, y1, yerr=yerr1, fmt='.', color='blue', capsize=5, label='Series 1')
plt.errorbar(x2, y2, yerr=yerr2, fmt='.', color='red', capsize=5, label='Series 2')
plt.errorbar(x3, y3, yerr=yerr3, fmt='.', color='pink', capsize=5, label='Series 3')
plt.errorbar(x4, y4, yerr=yerr4, fmt='.', color='green', capsize=5, label='Series 4')

plt.plot(x1, intercept1 + slope1*x1, color='blue', linestyle='--')
plt.plot(x2, intercept2 + slope2*x2, color='red', linestyle='--')
plt.plot(x3, intercept3 + slope3*x3, color='pink', linestyle='--')
plt.plot(x4, intercept4 + slope4*x4, color='green', linestyle='--')


plt.ylabel('ln(Count Rate / Sample)')
plt.xlabel('Sample')
plt.title('ln(Count Rate / Sample) vs. Sample with Linear Regression')
plt.legend()
plt.grid(True)


plt.show()

print("Shelf 1: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope1, intercept1, r_value1**2))
print("Shelf 2: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope2, intercept2, r_value2**2))
print("Shelf 3: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope3, intercept3, r_value3**2))
print("Shelf 4: y = {:.4f}x + {:.4f}, R^2 = {:.4f}".format(slope4, intercept4, r_value4**2))
print("U1=", std_err1)
print("U2=", std_err2)
print("U3=", std_err3)
print("U4=", std_err4)
```
